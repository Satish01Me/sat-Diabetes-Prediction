# Split independent variable and dependent variable

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1:].values
```

```python
x
```

```python
y
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x
```

# Split train and test dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

# Logistic Regression Model

Train Model

```python
from sklearn.linear_model import LogisticRegression
log_r=LogisticRegression()
log_r.fit(x_train,y_train)
```

Test Model

```python
pre=log_r.predict(x_test)
```

```python
np.concatenate((y_test.reshape(len(y_test),1),pre.reshape(len(pre),1)),1)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, pre)
print(cm)
accuracy_score(y_test, pre)
```

We goted the accuracy of our model is 87.8%

# K- nearest Neighbor Classification Model

Train Model

```python
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier(n_neighbors=7,metric='minkowski',p=2)
knc.fit(x_train,y_train)
```

Test Model

```python
pre_knc=knc.predict(x_test)
```

```python
np.concatenate((y_test.reshape(len(y_test),1),pre_knc.reshape(len(pre_knc),1)),1)
```

```python
cm = confusion_matrix(y_test, pre_knc)
print(cm)
accuracy_score(y_test, pre_knc)
```

We goted the accuracy of our model is 84.35%
